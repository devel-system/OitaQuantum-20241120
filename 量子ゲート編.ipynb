## **Step1 古典：量子回路を準備する**
量子ゲート方式の量子計算を実行するためには以下の流れで行います。

1.   量子ビットの数を設定して量子回路の初期状態を準備する。
2.   1で準備した量子回路に量子ビットゲートを作用させる。
3.   2で作成した量子回路を用いて量子計算を実行する。
4.   実行結果をもとに評価を行う。

Step1 では量子ビットの数を設定して量子回路の初期状態を準備してみます。

### **量子回路の初期状態の準備**
`qiskit` では `QuantumCircuit()` というのを用いることで量子回路の初期状態を準備することができます。 \
`QuantumCircuit()` の括弧の中に数字を入れることで量子ビットの数を指定することができます。

例 \
`QuantumCircuit(5)` → 5量子ビットの量子回路を準備。

```python
from qiskit import QuantumCircuit

# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

## **Step2 古典：準備した量子回路に量子ビットゲートを作用させる**
Step1で準備した量子回路に量子ビットゲートを作用させていきます。 \
この量子ビットゲートが量子計算の根幹部分にあたります。 \
主な量子ビットゲートは以下のようになります。

**1量子ビットゲート**

X, Y, Z, H, S, T, Rx, Ry, Rz, P, U

**2量子ビットゲート**

CX, CY, CZ, CH, SWAP, CRx, CRy, CRz, CP, CU

**3量子ビットゲート**

CCX



### **量子ビットゲートを作用させる**
1量子ビットゲートであるXゲートを作用させてみます。 \
`qiskit` では `x()` というのを用いることでXゲートを作用させることができます。 \
`x()` の括弧の中に数字を入れることで量子ビットゲートをどこに作用させるか指定することができます。

例 \
`x(1)` → 量子回路の上から1番目にxゲートを作用。

<font color="red">※ 作用させる場所は0番目から数えなければならない。</font>

```python
from qiskit import QuantumCircuit

# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子ビットゲートを作用
qc.x(0)

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

`qc.x()` は繰り返し記述することで複数の量子ビットゲートを作用させることができます。

```python
from qiskit import QuantumCircuit

# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子ビットゲートを作用
qc.x(0)
qc.x(1)

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

2量子ビットゲートの場合は2つ作用させる場所を指定する必要があります。 \
`cx()` というのを用いることでCXゲートを作用させることができます。

例 \
`cx(0, 1)` → 量子回路の上から0番目と1番目にcxゲートを作用。

```python
from qiskit import QuantumCircuit

# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子ビットゲートを作用
qc.cx(0, 1)

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

## **Step3 量子：作成した量子回路を用いて量子計算を実行する**
Step2で作成した量子回路に **観測** という量子ビットを読み取る操作を行うことで量子計算を実行することができます。 \
量子計算には実機はもちろんだが、シミュレータを用いて実行することもできます。 \
（観測の具体的な説明はここでは省略します。）

### **量子回路の観測する場所を指定する**
観測には量子ビットゲートの場合と同様に観測する場所を指定する必要があります。 \
今回は簡単に全ての量子ビットを読み取ってみます。

`qiskit` では `measure_all()` というのを用いることで観測に全ての量子ビットを指定することができます。

```python
from qiskit import QuantumCircuit

# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子ビットゲートを作用
qc.x(0)

# 観測する場所を指定
qc.measure_all()

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

### **量子計算を実行する**
作成した量子回路をデバイスに渡して、量子計算を実行してみます。 \
今回は [IBM Quantum](https://quantum.ibm.com/services/resources) に掲載している実機とシミュレータを使用できるようにしています。 \

`IBMQ_TOKEN` には [IBM Quantum](https://quantum.ibm.com/) から取得したトークンを入力してください。\
（トークンを入力しなかった場合はシミュレータが動きます。）

`IBMQ_DEVICE_NAME` には上の表を参考に実際に動かす量子コンピュータの名前を入力してください。

`qiskit` では実機の場合は `QiskitRuntimeService`、シミュレータの場合は `StatevectorSampler` というのを用いることで量子計算を実行することができます。

```python
from qiskit import QuantumCircuit
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import SamplerV2 as Sampler
from qiskit.primitives import StatevectorSampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

#######################
###    初期設定     ###
#######################
TOKEN = ""
BACKEND_NAME = "ibm_brisbane"
SHOTS = 1000

#######################
###  量子回路作成   ###
#######################
# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子ビットゲートを作用
qc.x(0)

# 観測する場所を指定
qc.measure_all()

#######################
###  量子計算実行   ###
#######################
if TOKEN:
    # QiskitRuntimeService
    service = QiskitRuntimeService(channel="ibm_quantum", token=TOKEN)
    backend = service.backend(name=BACKEND_NAME)
    pm = generate_preset_pass_manager(backend=backend, optimization_level=1)

    # Compile
    isa_circuit = pm.run(qc)
    sampler = Sampler(backend)

else:
    isa_circuit = qc
    sampler = StatevectorSampler()

job = sampler.run([isa_circuit], shots=SHOTS)
bitvals = job.result()[0].data.meas

if TOKEN:
    print(f">>> Job_id: {job.job_id()}")
    print(f">>> Backend: {backend}")
else:
    print(f">>> Backend: StatevectorSampler")
print(f">>> Counts: {bitvals.get_counts()}")

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

## **Step4 古典：実行結果をもとに評価を行う**
量子計算の実行結果をもとに評価を行います。 \
今回は得られた結果をグラフに表示してみましょう。

`qiskit` では描画ツールが含まれており、`plot_histogram` というのを用いることで実行結果を描画することができます。

```python
from qiskit.visualization import plot_histogram
counts = bitvals.get_counts()

# 実行結果の描画
plot_histogram(counts, title="Sampling Result")
```

## **実問題に適応させるためには**

この量子回路にある量子ビットゲートを作用させてみます。

```python
from qiskit import QuantumCircuit
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import SamplerV2 as Sampler
from qiskit.primitives import StatevectorSampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

#######################
###    初期設定     ###
#######################
TOKEN = ""
BACKEND_NAME = "ibm_brisbane"
SHOTS = 1000

#######################
###  量子回路作成   ###
#######################
# 量子回路の初期状態を準備
qc = QuantumCircuit(2)

# 量子ビットゲートを作用
qc.h(range(2))
qc.cz(0, 1)

qc.h(range(2))
qc.x(range(2))
qc.cz(0, 1)
qc.x(range(2))
qc.h(range(2))

# 観測する場所を指定
qc.measure_all()

#######################
###  量子計算実行   ###
#######################
if TOKEN:
    # QiskitRuntimeService
    service = QiskitRuntimeService(channel="ibm_quantum", token=TOKEN)
    backend = service.backend(name=BACKEND_NAME)
    pm = generate_preset_pass_manager(backend=backend, optimization_level=1)

    # Compile
    isa_circuit = pm.run(qc)
    sampler = Sampler(backend)

else:
    isa_circuit = qc
    sampler = StatevectorSampler()

job = sampler.run([isa_circuit], shots=SHOTS)
bitvals = job.result()[0].data.meas

if TOKEN:
    print(f">>> Job_id: {job.job_id()}")
    print(f">>> Backend: {backend}")
else:
    print(f">>> Backend: StatevectorSampler")
print(f">>> Counts: {bitvals.get_counts()}")

# 量子回路の描画
qc.draw(output="mpl", style="iqp")
```

```python
from qiskit.visualization import plot_histogram
counts = bitvals.get_counts()

# 実行結果の描画
plot_histogram(counts, title="Sampling Result")
```

4つの組み合わせだったのが、量子ビットゲートを作用させることで1つの結果のみが出力されました。
